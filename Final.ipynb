```python
model = YOLO('yolov8m.pt')

# Define the path to the dataset configuration file
data_yaml_path = "warp-d_config.yaml"

# Number of epochs to train for
epochs = 100


# Train the model with all enhancements
results = model.train(
    data=data_yaml_path,
    epochs=epochs,
    imgsz=640,
    device=0,


    degrees=25.0,
    translate=0.1,
    scale=0.4,
    shear=5.0,
    perspective=0.001,
    flipud=0.5,
    fliplr=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    mosaic=1.0,
    mixup=0.1
)
```
